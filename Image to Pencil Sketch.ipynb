<h2><font color="red">Importing required libraries: </font></h2>

```python
import cv2
import numpy as np
from IPython.display import display 
from PIL import Image
```

<h2><font color="red"> Displaying Real Image: </font></h2>

```python
image=cv2.imread('lion.jpg')
```

```python
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
display(Image.fromarray(image))
```

<h2><font color="red">Grayscale Image: </font></h2>

```python
gray_img=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_img))
```

<h2><font color="red">Negative Image: </font></h2>

```python
neg_img=cv2.bitwise_not(gray_img)
display(Image.fromarray(neg_img))
```

<h2><font color="red">Blurr Image: </font></h2>

```python
blur_img=cv2.GaussianBlur(neg_img,(21,21),0)
display(Image.fromarray(blur_img))
```

<h2><font color="red">Inverted Blur Image: </font></h2>

```python
inv_blur=cv2.bitwise_not(blur_img)
display(Image.fromarray(inv_blur))
```

<h2><font color="red">Finally pencil sketch Image: </font></h2>

```python
sketch=cv2.divide(gray_img,inv_blur,scale=190)
display(Image.fromarray(sketch))
```
